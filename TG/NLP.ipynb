```python
sample_string="In today's area of internet and online services, data is generating at incredible speed and amount. Generally, Data analyst, engineer, and scientists are handling relational or tabular data. These tabular data columns have either numerical or categorical data. Generated data has a variety of structures such as text, image, audio, and video. Online activities such as articles, website text, blog posts, social media posts are generating unstructured textual data. Corporate and business need to analyze textual data to understand customer activities, opinion, and feedback to successfully derive their business. To compete with big textual data, text analytics is evolving at a faster rate than ever before."

from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokenize(sample_string) # line by line or for dot 
```

```python
word_tokenize(sample_string) # word by word
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

words_in_sample_string = word_tokenize(sample_string)

stop_words = set(stopwords.words("english"))
#stop_words.update({';', ',', '.', ':'}) # for add stop word from the list
#stop_words.discard({'I', 'not'}) # for delet stop word from the list

filtered_list = []
for word in words_in_sample_string:
    if word.casefold() not in stop_words: # casefold => for lower case
         filtered_list.append(word)
            
filtered_list
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_list]
stemmed_words
```
